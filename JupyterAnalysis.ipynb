```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('weight-height.csv')

data
```

# **Population Statistics:**

```python
population_height_mean = round(data['Height'].mean(),3)
population_height_std = round(data['Height'].std(ddof=0),3)

print(f'Population height mean: {population_height_mean}','\n\n'
      f'Population height standard deviation: {population_height_std}')
```

```python
sns.histplot(data=data['Height'],bins=30,stat='density',color='beige')
sns.kdeplot(data=data['Height'],color='red')
plt.xlabel('Height')
plt.ylabel('Probability Density')
plt.title('Population Histogram of Height')
plt.axvline(population_height_mean,label='Population mean',linewidth=2)
plt.legend()
plt.show()
```

# **Sample Analysis For Mean:**

```python
# 100 samples of size n = 5

random_500 = np.array(data['Height'][np.random.randint(0,10000,500)]).reshape(100,5)

individual_means_of_100_samples_of_5 = random_500.mean(axis=1)
```

```python
mean_of_individual_means_of_5 = round(individual_means_of_100_samples_of_5.mean(),3)
std_of_individual_means_of_5 = round(individual_means_of_100_samples_of_5.std(ddof=1),3)
theo_std_5 = round(population_height_std/(5)**(1/2),3)

print(f'Sample height mean: {mean_of_individual_means_of_5}','\n\n'
      f'Sample height standard deviation: {std_of_individual_means_of_5}','\n\n'
      f'Theoritical Sample height standard deviation: {theo_std_5}')
```

```python
sns.histplot(data=individual_means_of_100_samples_of_5,bins=30,stat='density',color='beige')
sns.kdeplot(data=individual_means_of_100_samples_of_5,color='red')
plt.xlabel('Height')
plt.ylabel('Probability Density')
plt.title('Means of 100 Samples of size = 5 Histogram')
plt.axvline(population_height_mean,label='Population mean',linewidth=2)
plt.axvline(mean_of_individual_means_of_5,label='Sample mean',linewidth=2,color='green')
plt.legend()
plt.show()
```
